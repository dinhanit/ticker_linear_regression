# Task 4

```python
def Plot_Ticker(ticker):
    data = pd.read_csv('Task2/'+'df_'+ticker+'.csv')
    
    attributes = ['Open', 'High', 'Low', 'Close', 'Volume']
    plt.subplots(figsize=(20,10))

    for i, col in enumerate(attributes):
      plt.subplot(2,3,i+1) 
      sb.distplot(data[col])
    
    plt.subplots(figsize=(20,10))
    for i, col in enumerate(attributes):
      plt.subplot(2,3,i+1)
      sb.boxplot(data[col])
    plt.show()
    
    plt.figure(figsize=(15,5))
    plt.plot(data['Close'])
    plt.title('ANV close price.', fontsize=15)
    plt.ylabel('Price in VND.')
    plt.show()
```

```python
for ticker in tickers:
     Plot_Ticker(ticker)
```

```python
class TrainModel:
    def __init__(self,ticker):
        self.data = pd.read_csv('Task2/'+'df_'+ticker+'.csv')
        self.model = None
        self.Train()

    def Train(self):
        y= self.data['Close']
        x = self.data.drop(['Close','TradingDate'],axis=1)
        x = np.array(x)
        y = np.array(y)
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=0)
        reg = LinearRegression()
        reg.fit(x_train,y_train)
        print('Mean Square Error for data '+ticker+' = ',reg.score(x_test,y_test))
        self.model = reg
        
    def Predict(self,x_test):
        return list(self.model.predict(x_test))[0]
    def Data(self):
        return self.data
```
