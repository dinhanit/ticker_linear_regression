### Khám phá dữ liệu dùng một số thư viện Python

```python
# Thêm thư viện numpy đặt tên đại diện là np
# np thường được dùng để tính toán trên ma trận, mảng (array)
import numpy as np

# Thêm thư viện pandas đặt tên đại diện là pd
# pd thường được dùng để thao tác trên dữ liệu dạng bảng, time series,...
import pandas as pd

# Thêm thư viện matplotlib.pyplot đặt tên đại diện là plt
# plt thường được dùng đề vẽ biểu đồ (figures)
import matplotlib.pyplot as plt

# Thêm thư viện seaborn đặt tên đại diện là sb
# sb là một thư viện dựa trên matplotlib, 
# sb thường được dùng để trực quan dữ liệu với nhiều tiện lợi hơn, 'ít tốn công sức hơn' nếu dùng plt
import seaborn as sb

# Thêm thư viện warnings (cảnh báo) và tắt các cảnh báo
import warnings
warnings.filterwarnings('ignore')
```

```python
# Hiển thị biểu đồ histogram của mã chứng khoán ANV
attributes = ['Open', 'High', 'Low', 'Close', 'Volume']
plt.subplots(figsize=(20,10))
 
for i, col in enumerate(attributes):
  plt.subplot(2,3,i+1) # hiển thị đồ thị con (subplot) trên 2 dòng, mỗi dòng hiển thị 3 đồ thị
  sb.distplot(df_anv[col])
plt.show()
```

```python
# Hiển thị biểu đồ boxplot của mã chứng khoán ANV
plt.subplots(figsize=(20,10))
for i, col in enumerate(attributes):
  plt.subplot(2,3,i+1)
  sb.boxplot(df_anv[col])
plt.show()
```

```python
# Hiển thị đồ thị giá đóng cửa của mã chứng ANV (biến df_anv)
plt.figure(figsize=(15,5))
plt.plot(df_anv['Close'])
plt.title('ANV close price.', fontsize=15)
plt.ylabel('Price in VND.')
plt.show()
```
